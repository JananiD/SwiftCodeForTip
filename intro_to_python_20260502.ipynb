#### Using `print()` with string and variable concatenation

```python
num = 40
fruit = 'apple'
```

use the elements above to combine text with `print()` as: *"We have 40 apples"*

```python
#method 1 - least effective - requires adding spaces and converting int to string
print('We have'+' '+str(num)+' '+fruit)
```

```python
#method 2 - efficient no need for conversion - drawback: it inserts spaces for commas
print('We have',num,fruit)
```

```python
#drawback example - plural s has space
print('We have',num,fruit,'s')
```

```python
# method 3 - stands for format - Bassel's favorite
print(f'We have {num} {fruit}s')
```

```python
#method 4 - format at the end
print('We have {} {}s'.format(num, fruit)) #order matters unless you use indices
```

Bassel's recommendation: use for long text where you can edit variables all in one list.

```python
message = '''Hello,
{0} to our amazing Python {1}!
You'll {2} exciting content and great examples.
The students enjoyed the {1}.
'''.format('Welcome', 'class','find')

print(message)
```

> Hypothetically, imagine you have a very long text and a large list of variables - you can edit all of them in one location instead of finding them in the text. Also, same variables can be used in multiple places (e.g. class).

## Dictionaries

- A type of data structure with "key-value" pair
- It is mutable and unordered
- The values inside it can be any data type: list, int, float, etc...
- Use `{}` and `:` to define it

```python
# online auto store
my_data = {
    #'key':'value'
    'brand': 'Ford',
    'model': 'Mustang',
    'year': 1964, #int doesn't have to have quotes so we can preserve the int properties
    'color': 'red'
}

my_data
```

```python
# get the model
my_data['model']
```

```python
print(f"My car is {my_data['brand']} {my_data['model']}. It is {my_data['color']}.")
```

```python
# update a value in a dict
my_data['color'] = 'black'
my_data
```

```python
#remove a key. The value will automatically get deleted too
del my_data['color']
my_data
```

```python
my_data.keys()
```

```python
my_data.values()
```

To merge or add more elements:

```python
my_data_b = {'engine': 'V6', 'type': 'coupe'}

my_data.update(my_data_b) #for adding a dictionary/multiple items
my_data
```

```python
# the pop function called popitem
my_data.popitem()
my_data
```
